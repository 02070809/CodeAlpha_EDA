```python
import pandas as pd

url = "https://raw.githubusercontent.com/curran/data/gh-pages/superstoreSales/superstoreSales.csv"
df = pd.read_csv(url, encoding='latin1')

df.head()
```

```python
print("Total Rows:", df.shape[0])
print("Total Columns:", df.shape[1])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print("=== TOP SELLING CATEGORIES ===")
print(df['Product Category'].value_counts())

print("\n=== TOTAL SALES BY REGION ===")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False))

print("\n=== AVERAGE ORDER PROFIT ===")
print("Average Profit per Order:", round(df['Profit'].mean(), 2))

print("\n=== TOP 5 MOST PROFITABLE PRODUCTS ===")
print(df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head())
```
